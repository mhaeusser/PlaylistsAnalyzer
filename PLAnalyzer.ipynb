```python
# loads contents of a url
from urllib.request import urlopen 
import re
def read_url(url): 
    return urlopen(url).read().decode()
```

```python
def extract_raw_playlists(text):
    start = text.index('<!-- Start -->')
    end = text.index('<!-- End -->')
    raw = text[start+14 : end]
    return raw
```

```python
def replace_junk(text):
    text = text.replace('&amp;', '&')
    text = text.replace('&#8211;', '-')
    text = text.replace('&#8216;', "'")
    text = text.replace('&#8217;', "'")
    text = text.replace('- <i>Jingle</i>', '')
    return text
```

```python
def norm_linebreaks(text):
    #text = text.replace('</p>\n<p>', '<br/>')
    text = text.replace('<br />', '\n')
    text = text.replace('<br/>', '\n')
    text = text.replace('<p>', '')
    text = text.replace('</p>', '')
    text = text.replace('\n\n', '\n')
    return text
```

```python
def parse_date(line):
    # parses lines like 'Playlist 15.12.2003:' or 'Playlist 3.12.2004 (zu Gast: DJ Bernie (Rudeboy’s Delight)):'
    first_dot_index = line.index('.')
    second_dot_index = line.index('.', first_dot_index+1)
    #print(second_dot_index)
    date = line[9:second_dot_index+5]
    return date
```

```python
def parse_song_line(line):
    # parses lines like '- Lionsclub: The Persuaders'
    # or 'Victor Rice: Fique <img class="alignnone size-full wp-image-37" src="http://ska.one/new.gif"/>'
    line = line[2:]
    NEW_MARK = ' <img class="alignnone size-full wp-image-37" src="http://ska.one/new.gif"/>'
    SKA21_MARK = ' <a href="http://file-under-ska.de/old/ska21/" target="_blank" rel="noopener"><img src="http://ska.one/ska21_16.gif" alt="Track von SKA 21" border="0"/></a>'
    is_new = False
    if (line.endswith(NEW_MARK)):
        is_new = True
        line = line[0 : -len(NEW_MARK)]
    is_ska21 = False
    if (line.endswith(SKA21_MARK)):
        is_ska21 = True
        line = line[0 : -len(SKA21_MARK)]
    both = line.split(": ")
    artist = both[0]
    title = both[1]
    return artist, title, is_new, is_ska21
```

```python
def parse_url(url):
    text = read_url(url)
    text = extract_raw_playlists(text)
    text = replace_junk(text)
    text = norm_linebreaks(text)
    lines = text.split('\n')
    return lines
```

```python
def parse_lines(lines):
    songs = list()
    errors = list()
    for line in lines:
        #print("parsing: '" + line + "'")
        if not line:
            #print('Ignoring empty line')
            k=0 # noop
        elif (line.startswith('Playlist')):
            date = parse_date(line)
        elif (line.startswith('- ')):
            artist, title, is_new, is_ska21 = parse_song_line(line)
            songs.append([artist, title, date, is_new, is_ska21])
        else:
            #print("Cannot parse '" + line + '"')
            errors.append(line)
        #print('line: ' + artist + ' / ' + title + ' / ' + date)
    return songs, errors
```

```python
def parse_playlist(url, all_songs, all_errors):
    text = read_url(url)
    text = extract_raw_playlists(text)
    text = replace_junk(text)
    text = norm_linebreaks(text)
    lines = text.split('\n')
    songs, errors = parse_lines(lines)
    all_songs.append(songs)
    all_errors.append(errors)
```
