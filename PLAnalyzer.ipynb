```python
# loads contents of a url
from urllib.request import urlopen 
import re
def read_url(url): 
    return urlopen(url).read().decode()
```

```python
def extract_raw_playlists(text):
    start = text.index('<!-- Start -->')
    end = text.index('<!-- End -->')
    raw = text[start+14 : end]
    return raw
```

```python
def replace_junk(text):
    text = text.replace('&amp;', '&')
    text = text.replace('&#8211;', '-')
    text = text.replace('&#8216;', "'")
    text = text.replace('&#8217;', "'")
    text = text.replace('- <i>Jingle</i>', '')
    return text
```

```python
def replace_tags(text):
    text = text.replace(' <img class="alignnone size-full wp-image-37" src="http://ska.one/new.gif"/>', '<new/>')   
    text = text.replace(' <a href="http://file-under-ska.de/old/ska21/" target="_blank" rel="noopener"><img src="http://ska.one/ska21_16.gif" alt="Track von SKA 21" border="0"></a>', '<ska21/>')   
    return text
```

```python
def norm_linebreaks(text):
    #text = text.replace('</p>\n<p>', '<br/>')
    text = text.replace('<br />', '\n')
    text = text.replace('<br/>', '\n')
    text = text.replace('<p>', '')
    text = text.replace('</p>', '')
    text = text.replace('\n\n', '\n')
    return text
```

```python
def parse_song_line(line):
    # parses lines like - Lionsclub: The Persuaders'
    line = line[2:]
    both = line.split(": ")
    artist = both[0]
    title = both[1]
    return artist, title
```

```python
def parse_date(line):
    # parses lines like 'Playlist 15.12.2003:'
    date = line[9:-1]
    return date
```

```python
def parse_url(url):
    text = read_url(url)
    text = extract_raw_playlists(text)
    text = replace_junk(text)
    text = replace_tags(text)
    text = norm_linebreaks(text)
    lines = text.split('\n')
    return lines
```

```python
def parse_lines(lines):
    songs = list()
    date = 'none'
    artist = 'none'
    title = 'none'
    for line in lines:
        print('parsing: ' + line)
        if (line.startswith('Playlist')):
            date = parse_date(line)
        elif (line.startswith('- ')):
            artist, title = parse_song_line(line)
            # TODO add to list
            songs.append([artist, title, date])
        elif not line:
            print('Ignoring empty line')
        else:
            print('Cannot handle "' + line + '"')
        print('line: ' + artist + ' / ' + title + ' / ' + date)
    return songs
```

```python
def go():
    lines = parse_url('http://ska.one/playlists/playlists-2003/')
    songs = parse_lines(lines)
    return songs
```
