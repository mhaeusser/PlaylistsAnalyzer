```python
from urllib.request import urlopen 
import re
def read_url(url): 
    return re.sub('\\s+', ' ', urlopen(url).read().decode())
```

```python
playlist_2003_url = 'http://ska.one/playlists/playlists-2003/'
playlist_2003_text = read_url(playlist_2003_url)
```
