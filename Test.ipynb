```python
from urllib.request import urlopen 
import re
def read_url(url): 
    return re.sub('\\s+', ' ', urlopen(url).read().decode())
```

```python
p2004_url = 'http://ska.one/playlists/playlists-2004/'
p2004 = read_url(p2004_url)
```

```python
c2004 = p2004[p2004.index('<!-- Start -->')+19 : p2004.index('<!-- End -->')-4]
```
